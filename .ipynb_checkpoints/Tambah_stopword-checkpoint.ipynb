```python
from textblob import TextBlob
import tweepy
import pandas as pd
import numpy as py
import re, string
```

```python
#Key Token

mykeys = open('apikey.txt', 'r').read().splitlines()

api_key = mykeys[0]
api_key_secret = mykeys[1]
access_token = mykeys[2]
access_token_secret = mykeys[3]
auth_hendler = tweepy.OAuthHandler(api_key, api_key_secret)
auth_hendler.set_access_token(access_token, access_token_secret)
```

```python
#crawling data

api = tweepy.API(auth_hendler)

search_term = 'biznet'
tweet_amount = 300

#between two dates (just 1 month ago not more)
start_date = "2020-11-20"
end_date = "2020-11-30"

#Tweepy Cursor
tweets = tweepy.Cursor(api.search, q=search_term, lang='id', since=start_date, until=end_date).items(tweet_amount)

# Pulling information from tweets iterable 
tweets_ = [[tweet.created_at,tweet.user.screen_name, tweet.user.location, tweet.text] for tweet in tweets]

#Make DataFrame for tweets after crawling
tweets_list = pd.DataFrame(data=tweets_, columns=['Date','Username','Location','Text'])

# Creation of dataframe from tweets list
tweets_df = pd.DataFrame(tweets_list)
#tweets_df
```

```python
#convert data to 

tweets_df.to_csv('data.csv', index = False, header = True)
```

```python
#Preprocessing 

tweets = pd.read_csv('data.csv')
tweets.drop(['Date','Username','Location'], axis=1, inplace=True)
#tweets
```

```python
#step Cleansing
#remove user
def remove_pattern(tweet, pattern):
    r = re.findall(pattern, tweet)
    for i in r:
        tweet = re.sub(i, '', tweet)
    return tweet    
tweets['remove_user'] = py.vectorize(remove_pattern)(tweets['Text'], "@[\w]*")
```

```python
tweets
```

```python
def remove(tweet):
    #remove angka
    tweet = re.sub('[0-9]+', '', tweet)
    
    # remove stock market tickers like $GE
    tweet = re.sub(r'\$\w*', '', tweet)
 
    # remove old style retweet text "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    
    # remove hashtags
    # only removing the hash # sign from the word
    tweet = re.sub(r'#', '', tweet)
    
    #remove http
    tweet = re.sub(r'http\S+', '', tweet)
    
    return tweet
tweets['remove_RT'] = tweets['remove_user'].apply(lambda x: remove(x))
```

```python
tweets
```

```python
tweets.drop_duplicates(subset ="remove_user", keep = 'first', inplace = True)
```

```python
tweets.sort_values("remove_RT", inplace = True)
```
